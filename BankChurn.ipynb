Preprocess data -- Encode

```python
from sklearn import preprocessing
```

```python
labelencoder = preprocessing.LabelEncoder()
genenc = preprocessing.LabelEncoder()
edcenc=preprocessing.LabelEncoder()
marenc=preprocessing.LabelEncoder()
incenc=preprocessing.LabelEncoder()
cardenc=preprocessing.LabelEncoder()
```

```python
df_clean['Attrition_Flag'] = labelencoder.fit_transform(df_clean['Attrition_Flag'])
```

```python
df_clean['Gender'] = genenc.fit_transform(df_clean['Gender'])
df_clean['Education_Level'] = edcenc.fit_transform(df_clean['Education_Level'])
df_clean['Marital_Status'] = marenc.fit_transform(df_clean['Marital_Status'])
df_clean['Income_Category'] = incenc.fit_transform(df_clean['Income_Category'])
df_clean['Card_Category'] = cardenc.fit_transform(df_clean['Card_Category'])
```

```python
df_clean.head()
```

Train model--Using the customer demographics

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics
from sklearn.metrics import confusion_matrix  
```

```python
x = df_clean[['Gender','Education_Level','Marital_Status','Income_Category','Card_Category']]
y = df_clean['Attrition_Flag']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size =0.2,random_state=1)
```

```python
#Create object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(x_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(x_test)
```

Model Evaluation

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)  
```

```python
cm
```

```python
sns.heatmap(cm, annot=True, fmt='g')
```

```python
report = metrics.classification_report(y_test, y_pred)
```

```python
print(report)
```
